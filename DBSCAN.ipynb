```python
data = customer[["Fresh", "Frozen"]]
```

```python
customer = customer[["Grocery", "Milk"]]
#customer = customer.as_matrix().astype("float32", copy = False)
```

# 6. Scale the data

```python
stscaler = StandardScaler()

customer = stscaler.fit_transform(customer)
```

```python
#customer
data =stscaler.fit_transform(data)
```

# 7. Create and compute the model

```python
######## You will construct a DBSCAN object that requires a minimum of 15 
#data points in a neighborhood of radius 0.5 to be considered a core point.
dbsc = DBSCAN(eps = 0.5, min_samples = 15)
secondModel =DBSCAN(eps=0.5,min_samples=15)
```

```python
# fitting the mdoel
model=dbsc.fit(customer)
models =secondModel.fit(data)
```

```python
## we can extract our cluster labels and outliers to plot our results
labels = model.labels_
core_samples = np.zeros_like(labels, dtype = bool)
core_samples[model.core_sample_indices_] = True
core_samples_mask = np.zeros_like(model.labels_, dtype=bool)
core_samples_mask[model.core_sample_indices_] = True


## for second model
labelss = models.labels_
core_samples = np.zeros_like(labels, dtype = bool)
core_samples[models.core_sample_indices_] = True
core_samples_mask = np.zeros_like(model.labels_, dtype=bool)
core_samples_mask[model.core_sample_indices_] = True
```

```python
# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)


### for second model
# Number of clusters in labels, ignoring noise if present.
n_clusterss_ = len(set(labelss)) - (1 if -1 in labelss else 0)
n_noises_ = list(labelss).count(-1)
```

```python
print('Estimated number of clusters: %d' % n_clusters_)
print('Estimated number of noise points: %d' % n_noise_)



print('Estimated number of clusters: %d' % n_clusterss_)
print('Estimated number of noise points: %d' % n_noises_)
```

# 8. Test the performance of DBSCAN

```python
## the result of the accuracy
score=metrics.silhouette_score(customer,labels)
print('Percentage of accuracy: %f' % score)

##  for second  model
scores=metrics.silhouette_score(customer,labelss)
print('Percentage of accuracy: %f' % scores)
```

# 9. Visualization 

```python
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = customer[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = customer[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.xlabel("Groceries")
plt.ylabel("Milk")
plt.show()
```

```python
unique_labels = set(labelss)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labelss == k)

    xy = data[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = data[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusterss_)
plt.xlabel("Fresh")
plt.ylabel("Frozen")
plt.show()
```
